```r
library(tidyverse)
```

```r
covid_data <- read.csv("https://opendata.ecdc.europa.eu/covid19/virusvariant/csv/data.csv", header = TRUE)
covid_data
```

```r
covid_data_step0 <- covid_data |>
    select(country, year_week, new_cases, variant) 
covid_data_step0
```

```r
#find when Omicron get start
covid_data_step1 <- covid_data_step0 |>
    filter(variant == "B.1.1.529" |variant == "BA.1" |variant ==  "BA.1.1" |variant ==  "BA.2" 
           |variant ==  "BA.3" |variant ==  "BA.4" |variant ==  "BA.5", new_cases != 0)
covid_data_step1
```

```r
covid_data_step2 <- covid_data_step0 |>
     filter(substr(year_week, 1, 4) == "2021" | substr(year_week, 1, 4) == "2022")
covid_data_step2
```

```r
#find top 10 countries
covid_data_step3 <- covid_data_step2 |>
    group_by(country) |>
    summarise(total_cases_year = sum(new_cases)) |>
    arrange(-total_cases_year) |>
    slice(0:10)
covid_data_step3

#SO we just use the data from first 10 countries in Europe
```

```r
covid_data_top10 <- covid_data_step2 |>
    filter(country == "France" | country == "Germany"| country == "Spain" |country == "Italy" |country == "Poland" |
         country == "Netherlands" |country == "Czechia" |country == "Belgium" |country == "Greece" |country == "Romania")
covid_data_top10
```

```r
most_variarant <- covid_data_top10 |>
    select(new_cases, variant) |>
    group_by(variant) |>
    summarize(total_new_cases = sum(new_cases)) |>
    arrange(-total_new_cases)
most_variarant
```

```r
# rename the B.1.1.529, BA.1, BA.1.1, BA.2, BA.3, BA.4 and BA.5  to Omicron
# rename the B.1.617.2 to Delta
covid_data_step4 <- covid_data_top10 |>
     mutate(variant=ifelse(variant == "B.1.1.529" |variant == "BA.1"|variant == "BA.1.1"|variant == "BA.2"
                           |variant == "BA.3"|variant == "BA.4"|variant == "BA.5","Omicron",variant))
covid_data_step5 <- covid_data_step4 |>
    mutate(variant=ifelse(variant == "B.1.617.2","Delta",variant))
covid_data_step5
```

```r
#rename the other variant into Other
covid_data_step6 <- covid_data_step5 |>
     mutate(variant=ifelse(variant != "Omicron" & variant != "Delta", "other",variant))
covid_data_step6
```

```r
covid_data_step7 <- covid_data_step6 |>
    group_by(country, year_week, variant) |>
    summarize(total_new_cases_aweek = sum(new_cases)) 
covid_data_step7
```

```r
options(repr.plot.width = 10, repr.plot.height = 10)
covid_data_scrapter <-  covid_data_step7 |>
    ggplot(aes(y = year_week, x = total_new_cases_aweek, color = variant, shape = variant)) +
        geom_line() +
        labs(y = "Time", x = "New cases aweek", color = "variant") 
        
        
covid_data_scrapter
```
